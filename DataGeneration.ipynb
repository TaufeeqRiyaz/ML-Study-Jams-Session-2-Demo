# Univariate Linear Regression Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Initialise the line and randomness

```python
np.random.seed(42)
m = 0.4
c = 6
start = 50
stop = 500
multiplier = 3
x = np.linspace(start, stop, multiplier*stop-start)
y = m*x + c + np.random.normal(size=multiplier*stop-start)*25
```

Plot the generated data

```python
plt.figure(figsize=(15, 10))
plt.scatter(x, y)
plt.show()
```

Convert this to dataframe and csv

```python
df = pd.DataFrame({'x1': x, 'y': y})
df.head()
```

```python
df.to_csv('UniLinRegData.csv', index = None, header=True)
```
